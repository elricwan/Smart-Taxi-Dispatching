```python
# set the coordinate system to 'epsg:4326', which is latitude and longitude coordinates.
import geopandas as gpd
coord_system = {'init': 'epsg:4326'}
shapefile = gpd.read_file("taxi_zones.shp").to_crs(coord_system)
shapefile = shapefile.drop(['Shape_Area', 'Shape_Leng', 'OBJECTID'], axis=1)
shapefile.head()
```

```python
my_list = [128, 107, 4, 261, 262, 263, 137, 209, 140, 13, 142, 143, 144, 24, 148, 151, 152, 153, 158, 161, 162, 163, 164, 166, 113, 41, 170, 43, 45, 125, 48, 50, 244, 186, 74, 194, 68, 12, 202, 75, 79, 141, 211, 87, 88, 90, 224, 100, 229, 230, 231, 232, 233, 234, 103, 236, 237, 238, 239, 104, 114, 243, 116, 246, 105, 120, 249, 42, 127]
```

```python
# note multiple polygon
#shapefile['geometry'][i-1].geoms[1].exterior.coords.xy[0]
import matplotlib.pyplot as plt
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5, forward=True)
for i in my_list:
    if i not in [153,194,202,103,104,105]:
        shape = shapefile['geometry'][i-1]
        try:
            points = shape.exterior.coords.xy
        except:
            if i in [127,74,262,233]:# multipolygon problem
                points = shape.geoms[0].exterior.coords.xy
            else:  # multipolygon problem  
                points = shape.geoms[1].exterior.coords.xy
        x = [i for i in list(points[0])]
        y = [i for i in list(points[1])]
        plt.plot(x,y)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.collections import PatchCollection
import numpy as np
from matplotlib.patches import Polygon
```

```python
ptchs=[]
colors = []
for i in my_list:
    if i not in [153,194,202,103,104,105]:
        shape = shapefile['geometry'][i-1]
        try:
            points = shape.exterior.coords.xy
            ptchs.append(Polygon(np.transpose(np.array(points))))
        except:
            if i in [127,74,262,233]:# multipolygon problem
                points = shape.geoms[0].exterior.coords.xy
                ptchs.append(Polygon(np.transpose(np.array(points))))
            else:  # multipolygon problem  
                points = shape.geoms[1].exterior.coords.xy
                ptchs.append(Polygon(np.transpose(np.array(points))))
        a = Demand[i]/max(Demand.values())
        b = 100*a 
        colors.append(b)
```

```python
colors = np.array(colors)
```

```python
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5, forward=True)
ax = fig.add_subplot(111)
p = PatchCollection(ptchs, cmap=matplotlib.cm.jet, alpha=0.4)
p.set_array(colors)
ax.add_collection(p)
plt.colorbar(p)
for i in my_list:
    if i not in [153,194,202,103,104,105]:
        shape = shapefile['geometry'][i-1]
        try:
            points = shape.centroid.xy
            plt.text(points[0][0],points[1][0],str(i))
        except:
            points = shape.geoms[1].centroid.xy
            plt.text(points[0][0],points[1][0],str(i))
ax.set_xlim(-74.0179, -73.9112)
ax.set_ylim(40.7016, 40.8780)
plt.show()
```
